```python
#Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.
#import numpy as np
#np.random.seed(3)
#x = 3 + np.random.normal(0, 2, 50)
#y = 3 + np.random.normal(0, 2, len(x))
#Note: Also add title, xlabel, and ylabel to the plot.

'''
A scatter plot is a type of plot that displays the relationship between two numerical variables. Each data point is represented as a marker, with the x-coordinate 
representing one variable and the y-coordinate representing the other variable. Scatter plots are useful for visualizing patterns, trends, or correlations between the
variables.
'''
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y)  # Create the scatter plot
plt.title('Scatter Plot')  # Add title
plt.xlabel('X')  # Add x-axis label
plt.ylabel('Y')  # Add y-axis label

plt.show()  # Display the plot
```

```python
#Q3: Why is the subplot() function used? Draw four line plots using the subplot() function. Use the following data:
#import numpy as np
#For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
#For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
#For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
#For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

'''
The `subplot()` function in Matplotlib is used to create a grid of subplots within a single figure. It allows you to plot multiple plots in a single figure, which can
be useful for comparing different datasets or visualizing multiple aspects of the same data.

Each subplot is organized in rows and columns, and you can specify the position of each subplot using the row, column, and index parameters of the `subplot()` function.
'''


import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 1, 2, 3, 4, 5])  # Common x-values for all lines

# Data for line 1
y1 = np.array([0, 100, 200, 300, 400, 500])

# Data for line 2
y2 = np.array([50, 20, 40, 20, 60, 70])

# Data for line 3
y3 = np.array([10, 20, 30, 40, 50, 60])

# Data for line 4
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a 2x2 grid of subplots
plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title('Line 1')

plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title('Line 2')

plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title('Line 3')

plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title('Line 4')

plt.tight_layout()  # Improve spacing between subplots

plt.show()  # Display the plot
```

```python
#Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
#import numpy as np
#company = np.array(["Apple", "Microsoft", "Google", "AMD"])
#profit = np.array([3000, 8000, 1000, 10000])

'''
A bar plot, also known as a bar chart, is a type of plot that uses rectangular bars to represent categorical data. It displays the values of different categories as
bars of varying heights or lengths. The length or height of each bar represents the magnitude or frequency of the data being represented.

Bar plots are commonly used to compare and visualize the values or frequencies of different categories. They are particularly useful for highlighting differences or 
similarities between categories, identifying the highest or lowest values, and showing the distribution or composition of data across categories.
'''

import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Bar plot
plt.figure(figsize=(8, 5))  # Adjust the figure size if needed
plt.bar(company, profit)
plt.title('Company Profit')
plt.xlabel('Company')
plt.ylabel('Profit')

plt.show()

# Horizontal bar plot
plt.figure(figsize=(8, 5))  # Adjust the figure size if needed
plt.barh(company, profit)
plt.title('Company Profit')
plt.xlabel('Profit')
plt.ylabel('Company')

plt.show()
```

```python
#Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
#box1 = np.random.normal(100, 10, 200)
#box2 = np.random.normal(90, 20, 200)

'''
A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset. It displays the five-number summary of a dataset,
which includes the minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum. The plot consists of a box that represents the interquartile range 
(IQR), with a line or "whisker" extending from each end of the box to indicate the range of the data. Additionally, individual data points outside the whiskers may be
shown as outliers.

Box plots are used to visually summarize and compare the distribution of numerical data across different categories or groups. They provide insights into the central
tendency, spread, and skewness of the data, as well as identify potential outliers. Box plots are especially useful when dealing with large datasets or when comparing
multiple datasets.
'''

import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]

plt.figure(figsize=(8, 5))  # Adjust the figure size if needed
plt.boxplot(data)
plt.title('Box Plot')
plt.xlabel('Box')
plt.ylabel('Values')

plt.xticks([1, 2], ['Box 1', 'Box 2'])  # Customize x-axis tick labels

plt.show()
```
