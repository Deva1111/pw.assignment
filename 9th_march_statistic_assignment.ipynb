```python
#Q1: What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? Explain with an example.
'''
The Probability Mass Function (PMF) and Probability Density Function (PDF) are both concepts used in probability theory and statistics to describe the probability 
distribution of a random variable.

1. Probability Mass Function (PMF):
The PMF is used to describe the probability distribution of a discrete random variable. It assigns probabilities to each possible value that the random variable can
take. The PMF gives the probability of observing a specific value of the random variable.

For example, let's consider a fair six-sided die. The random variable X represents the outcome of a single roll of the die. The PMF of X would be:
PMF(X = 1) = 1/6
PMF(X = 2) = 1/6
PMF(X = 3) = 1/6
PMF(X = 4) = 1/6
PMF(X = 5) = 1/6
PMF(X = 6) = 1/6

In this case, each outcome has an equal probability of 1/6, as the die is fair. The PMF assigns a probability to each possible value of X, satisfying the properties 
that the probabilities are non-negative and sum up to 1.

2. Probability Density Function (PDF):
The PDF is used to describe the probability distribution of a continuous random variable. Unlike the PMF, which assigns probabilities to individual values, the PDF 
gives the probability density at each point in the range of the random variable.

For example, let's consider a continuous random variable Y representing the heights of adult males. The PDF of Y would be a function that provides the relative 
likelihood of observing a specific height.

The PDF might have a shape resembling a bell curve, such as the normal distribution. It could be described by a function like the Gaussian distribution:
PDF(Y = y) = (1/σ√(2π)) * exp(-(y-μ)²/(2σ²))

In this case, μ represents the mean height of adult males, σ represents the standard deviation, and exp is the exponential function. The PDF does not directly give 
the probability of observing a specific height, but rather provides information about the density of probabilities over the range of possible heights. The probability
of observing a specific height within a certain range can be obtained by integrating the PDF over that range.

Both the PMF and PDF are important tools in probability and statistics for understanding the behavior and characteristics of random variables, whether discrete or
continuous.
'''
pass
```

```python
#Q5: What is Bernaulli Distribution? Give an Example. What is the difference between Bernoulli Distribution and Binomial Distribution?

'''
The Bernoulli distribution is a discrete probability distribution that models a single binary (yes/no or success/failure) random variable. It represents a situation
where there are only two possible outcomes, each with a fixed probability of occurrence.

In the Bernoulli distribution, there is a single parameter, usually denoted as p, which represents the probability of the "success" outcome (often labeled as 1). 
The probability of the "failure" outcome (labeled as 0) is given by q = 1 - p.

Mathematically, the Bernoulli distribution is defined as follows:

P(X = 1) = p
P(X = 0) = q = 1 - p

Here's an example to illustrate the Bernoulli distribution:

Consider a coin toss experiment, where the random variable X represents the outcome of a single coin flip. Let's assume that "heads" is considered a success (1) and
"tails" is considered a failure (0). The probability of getting heads is denoted as p. If the coin is fair, p = 0.5.

In this case, the Bernoulli distribution would be:
P(X = 1) = p = 0.5
P(X = 0) = q = 1 - p = 1 - 0.5 = 0.5

The difference between the Bernoulli distribution and the Binomial distribution lies in the number of trials or repetitions involved:

1. Bernoulli Distribution: Represents a single trial or experiment with two possible outcomes (success or failure). It models the probability of observing a single 
   success or failure.

2. Binomial Distribution: Represents the number of successes (or failures) in a fixed number of independent and identical Bernoulli trials. It models the probability
   of obtaining a specific number of successes in multiple trials.

The Binomial distribution has two parameters: n (the number of trials) and p (the probability of success in each trial). It is given by the formula:

P(X = k) = C(n, k) * p^k * q^(n-k)

where C(n, k) represents the number of combinations of n items taken k at a time.

In essence, the Bernoulli distribution is a special case of the Binomial distribution, where the number of trials (n) is fixed at 1. The Binomial distribution allows
for the analysis of multiple trials and the calculation of probabilities associated with different numbers of successes or failures.
'''
pass
```

```python
#Q7: Explain uniform Distribution with an example.

'''
The uniform distribution is a continuous probability distribution that represents a situation where all outcomes within a given interval are equally likely. In other 
words, it describes a random variable that has a constant probability density function (PDF) over a defined range.

Mathematically, the PDF of a uniform distribution is defined as:

f(x) = 1 / (b - a),   for a ≤ x ≤ b
       0,             otherwise

where 'a' and 'b' are the lower and upper bounds of the interval, respectively.

Here's an example to illustrate the uniform distribution:

Consider a fair six-sided die. The random variable X represents the outcome of a single roll of the die. The possible outcomes are numbers 1 to 6. In this case, each
outcome has an equal probability of occurring, making it a uniform distribution.

The PDF of the uniform distribution for this example would be:

f(x) = 1 / 6,   for 1 ≤ x ≤ 6
       0,        otherwise

The probability density function is constant within the interval [1, 6], and it is zero outside that interval.

In the uniform distribution, every possible value within the interval has the same probability of occurring. This is in contrast to other distributions, such as the 
normal distribution, where certain values are more likely to occur than others.

The uniform distribution is often used in situations where there is no inherent bias or preference for any specific value within the interval. It can be useful for 
modeling situations such as random number generation, lotteries, and some forms of sampling.
'''
pass
```

```python
#Q9: What is Central Limit Theorem? State the significance of the Central Limit Theorem.

'''
The Central Limit Theorem (CLT) is a fundamental concept in statistics that states that when independent and identically distributed random variables are summed or
averaged, regardless of their underlying distribution, the distribution of the sum or average tends to approach a normal distribution as the number of variables increases.

The Central Limit Theorem has several significant implications:

1. Normal Approximation: The CLT allows for the approximation of many real-world phenomena by a normal distribution. Even if the individual variables do not follow a normal
   distribution, the sum or average of a large number of them tends to exhibit a bell-shaped, symmetric distribution. This makes the normal distribution a widely applicable
   and convenient model in various statistical analyses.

2. Robustness to Non-Normality: The CLT provides reassurance that even if the underlying distribution of a variable is not known or is not normal, we can still rely on the 
   normal distribution as an approximation for the distribution of the sample mean. This is particularly useful when performing inference or constructing confidence intervals
   based on the sample mean.

3. Sample Size Independence: The CLT applies regardless of the shape of the population distribution. It is not contingent on the size or skewness of the population distribution. 
   As long as the random variables are independent and identically distributed, the distribution of the sample mean will converge to a normal distribution with increasing sample size.

4. Inference and Hypothesis Testing: The CLT is crucial for making inferences about population parameters and conducting hypothesis tests. It enables the use of parametric tests, such
   as t-tests and z-tests, which rely on the assumption of normality or the application of the normal distribution as an approximation.

5. Sample Mean Stability: The CLT provides stability to the sample mean. Regardless of the population distribution, the distribution of the sample mean tends to become more concentrated
   around the population mean as the sample size increases. This implies that larger samples provide more precise estimates of the population mean.

6. Population Variance Estimation: The CLT is also used in estimating the population variance. It states that the sample variance divided by the sample size minus one follows a chi-square
   distribution, which facilitates the calculation of confidence intervals for the population variance.

The Central Limit Theorem is a fundamental concept in statistics that enables researchers to make reliable inferences, approximate real-world phenomena, and apply a range of statistical 
techniques even when the underlying distributions are unknown or non-normal. It forms the basis for many statistical methods and has extensive applications in fields such as hypothesis testing,
confidence interval estimation, and sampling theory.
'''
pass
```
