```python
#Q1. What is the Probability density function?

'''
The probability density function (PDF) is a concept in probability theory and statistics. It describes the probability distribution of a continuous random variable. 
In simple terms, the PDF gives the probability of a random variable taking on a specific value or falling within a certain range of values.

The PDF is defined mathematically as a function that assigns probabilities to different outcomes. For a continuous random variable, the PDF is typically denoted as 
f(x), where x represents the variable's possible values. The PDF satisfies the following properties:

1. Non-negativity: The PDF is always non-negative, meaning that f(x) ≥ 0 for all x.

2. Integration over the entire range: The total area under the PDF curve over the entire range of possible values of x is equal to 1. Mathematically, it can be expressed
as ∫f(x) dx = 1, where the integral is taken over all possible values of x.

3. Probability interpretation: The probability of a random variable falling within a specific interval [a, b] is given by the integral of the PDF over that interval.
Mathematically, it is expressed as P(a ≤ x ≤ b) = ∫[a,b] f(x) dx, where [a, b] represents the interval of interest.

The shape of the PDF curve provides insights into the distribution of the random variable. Different probability distributions have distinct PDF shapes, such as the 
bell-shaped curve of the normal distribution or the skewed shape of the exponential distribution. By analyzing the PDF, one can make inferences about the likelihood 
of different values occurring for the random variable.

It's important to note that the PDF is applicable to continuous random variables. For discrete random variables, the concept of probability mass function (PMF) is 
used instead.
'''
pass
```

```python
#Q2. What are the types of Probability distribution?

'''
There are several types of probability distributions in statistics. Here are some commonly encountered distributions:

1. Normal Distribution (Gaussian Distribution): The normal distribution is a symmetric bell-shaped distribution. Many natural phenomena follow this distribution, and 
   it is often used as an approximation for various real-world processes. The distribution is defined by its mean and standard deviation.

2. Binomial Distribution: The binomial distribution describes the number of successes in a fixed number of independent Bernoulli trials. Each trial has two possible
   outcomes (success or failure) with a constant probability of success. The distribution is characterized by two parameters: the number of trials and the probability 
   of success.

3. Poisson Distribution: The Poisson distribution models the number of events that occur within a fixed interval of time or space. It is often used to describe rare 
   events. The distribution is characterized by a single parameter, λ (lambda), which represents the average rate of occurrence of the events.

4. Exponential Distribution: The exponential distribution models the time between events in a Poisson process, where events occur continuously and independently at a
   constant rate. It is often used to model the lifespan of products, waiting times, or time between occurrences of rare events. The distribution is characterized by a
   parameter called the rate parameter, which represents the average rate of occurrence.

5. Uniform Distribution: The uniform distribution has a constant probability density over a fixed interval. It implies that all values within the interval have an equal
   likelihood of occurring.

6. Gamma Distribution: The gamma distribution is a versatile distribution that extends the exponential distribution. It is often used to model waiting times, income, and
   various other positive-valued random variables. The distribution is characterized by two parameters: a shape parameter and a rate parameter.

7. Beta Distribution: The beta distribution is a continuous probability distribution defined on the interval [0, 1]. It is commonly used to model proportions or probabilities.
   The distribution is characterized by two shape parameters.

8. Student's t-Distribution: The t-distribution is used for inference when the sample size is small or the population standard deviation is unknown. It is similar in shape
   to the normal distribution but has heavier tails.

9. Chi-Square Distribution: The chi-square distribution arises in hypothesis testing and estimation involving the variance of a normal distribution. It is often used in 
   goodness-of-fit tests and in calculating confidence intervals for the variance.

These are just a few examples of probability distributions. There are many more distributions, each with its own characteristics and applications. The choice of distribution
depends on the nature of the data and the specific problem at hand.
'''
pass
```

```python
#Q3. Write a Python function to calculate the probability density function of a normal distribution with given mean and standard deviation at a given point.

import math

def normal_pdf(x, mean, std_dev):
    exponent = -(x - mean) ** 2 / (2 * std_dev ** 2)
    coefficient = 1 / (std_dev * math.sqrt(2 * math.pi))
    pdf = coefficient * math.exp(exponent)
    return pdf

mean = 0
std_dev = 1
x = 1.5
pdf_value = normal_pdf(x, mean, std_dev)
print(f"The PDF value at x={x} is {pdf_value:.4f}")

```

```python
#Q4. What are the properties of Binomial distribution? Give two examples of events where binomial distribution can be applied.

'''
The binomial distribution is a discrete probability distribution that describes the number of successes in a fixed number of independent Bernoulli trials. Here are the
properties of the binomial distribution:

1. Fixed Number of Trials: The binomial distribution considers a fixed number of trials, denoted as n. Each trial is independent and has only two possible outcomes: 
   success or failure.

2. Independent Trials: The trials in a binomial distribution are assumed to be independent, meaning that the outcome of one trial does not affect the outcome of another.

3. Constant Probability of Success: The probability of success in each trial is denoted as p and remains constant for all trials. The probability of failure is 1 - p.

4. Discrete Distribution: The binomial distribution is a discrete distribution, meaning that it describes the probabilities of obtaining a specific number of successes,
   ranging from 0 to n.

5. Probability Mass Function (PMF): The binomial distribution is characterized by its probability mass function, which gives the probability of obtaining k successes in
   n trials. The PMF of the binomial distribution is given by the formula:

   P(X = k) = (n choose k) * p^k * (1 - p)^(n - k)

   Where:
   - P(X = k) is the probability of getting k successes.
   - (n choose k) is the binomial coefficient, which represents the number of ways to choose k successes out of n trials.
   - p^k represents the probability of getting k successes.
   - (1 - p)^(n - k) represents the probability of getting n - k failures.

Two examples where the binomial distribution can be applied are:

1. Coin Flips: Consider a scenario where you flip a fair coin 10 times and count the number of heads. Each coin flip is a Bernoulli trial with a 50% chance of success
   (getting a head). The total number of heads in 10 flips follows a binomial distribution with parameters n = 10 (number of trials) and p = 0.5 (probability of success).

2. Product Defects: Suppose a manufacturing process produces electronic components, and the probability of a component being defective is 0.05. If we randomly select 
   100 components and count the number of defective ones, the number of defects follows a binomial distribution with n = 100 (number of trials) and p = 0.05 
   (probability of success, i.e., being defective).

In both examples, we have a fixed number of independent trials with a constant probability of success, making them suitable for modeling using the binomial distribution.
'''
pass
```

```python
#Q6. Write a Python function to calculate the cumulative distribution function of a Poisson distribution with given mean at a given point.

import math

def poisson_cdf(x, mean):
    cdf = 0.0
    for k in range(x + 1):
        cdf += (math.exp(-mean) * mean**k) / math.factorial(k)
    return cdf

mean = 3.5
x = 2
cdf_value = poisson_cdf(x, mean)
print(f"The CDF value at x={x} is {cdf_value:.4f}")
```

```python
#Q7. How Binomial distribution different from Poisson distribution?

'''
The binomial distribution and the Poisson distribution are both probability distributions that describe the number of events or successes. However, there are some key 
differences between the two:

1. Nature of Events: The binomial distribution is used when the number of events or trials is fixed and known in advance, with each event having only two possible outcomes
  (success or failure). In contrast, the Poisson distribution is used when the number of events is random or unknown, and the events occur in a continuous or infinite time
   or space.

2. Number of Trials: In the binomial distribution, the number of trials, denoted as 'n', is a fixed and finite number. Each trial is independent of others and has a constant
   probability of success. In the Poisson distribution, there is no fixed number of trials; instead, the distribution describes the number of events that occur within a 
   specific time or space.

3. Probability of Success: In the binomial distribution, the probability of success in each trial, denoted as 'p', remains constant across all trials. In the Poisson 
   distribution, the mean or average number of events, denoted as 'λ' (lambda), is used to describe the intensity or rate at which events occur. The probability of 
   individual events may vary, but the average rate remains constant.

4. Discrete vs. Continuous: The binomial distribution is a discrete distribution since it deals with a finite number of possible outcomes (e.g., 0, 1, 2, ..., n). 
   The Poisson distribution, on the other hand, is a discrete distribution that describes the number of events occurring in a continuous range or over time (e.g., 0, 1, 2, ...).

5. Shape of Distribution: The binomial distribution can be symmetric or skewed, depending on the probability of success and the number of trials. It approaches a symmetric 
   bell-shaped distribution as the number of trials increases. The Poisson distribution is often positively skewed (skewed to the right) due to the nature of rare events 
   occurring with low probabilities.

6. Approximation: In certain cases, the Poisson distribution can be used as an approximation of the binomial distribution under specific conditions. This occurs when the number 
   of trials is large (n → ∞) and the probability of success is small (p → 0), with the product np approaching a finite value. In such cases, the Poisson distribution with 
   λ = np provides a good approximation to the binomial distribution.

Overall, the choice between using a binomial distribution or a Poisson distribution depends on the specific characteristics of the events or trials being modeled, such as whether
the number of trials is fixed or random and the nature of the probability of success or event occurrence.
'''
pass
```

```python
#Q8. Generate a random sample of size 1000 from a Poisson distribution with mean 5 and calculate the sample mean and variance.


import numpy as np

# Parameters
mean = 5  # Mean of the Poisson distribution
sample_size = 1000  # Sample size

# Generate random sample
sample = np.random.poisson(mean, size=sample_size)

# Calculate sample mean and variance
sample_mean = np.mean(sample)
sample_variance = np.var(sample)

print(f"Sample Mean: {sample_mean:.4f}")
print(f"Sample Variance: {sample_variance:.4f}")
```

```python
#Q9. How mean and variance are related in Binomial distribution and Poisson distribution?

'''
In both the binomial distribution and the Poisson distribution, the mean and variance are related, but the specific relationship differs between the two distributions.

Binomial Distribution:
For a binomial distribution, the mean (μ) and variance (σ^2) are related as follows:
μ = n * p
σ^2 = n * p * (1 - p)
where:
- n is the number of trials
- p is the probability of success in each trial

In the binomial distribution, the mean is equal to the number of trials (n) multiplied by the probability of success (p). The variance is the product of the number of
trials, the probability of success, and the probability of failure (1 - p).

Poisson Distribution:
For a Poisson distribution, the mean (λ) and variance (σ^2) are equal:
μ = λ
σ^2 = λ
where:
- λ (lambda) is the average rate or intensity of events occurring in a specific time or space.

In the Poisson distribution, both the mean and the variance are equal to the average rate of events (λ). This means that the mean and variance have the same value in 
a Poisson distribution, indicating a direct relationship between the two.

The relationship between mean and variance in both distributions reflects the characteristics of the respective distributions. In the binomial distribution, as the 
probability of success varies (p), it affects both the mean and the variance. In the Poisson distribution, the average rate (λ) determines both the mean and the variance,
indicating a constant spread around the mean.
'''
pass
```
