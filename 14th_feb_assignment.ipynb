```python
#Q4. Write a python program to create two threads. Thread one must print the list of squares and thread two must print the list of cubes

import threading

# Define a function to print the list of square
def print_squares():
    for i in range(1, 11):
        print(f"{i} squared is {i**2}")

# Define a function to print the list of cubes
def print_cubes():
    for i in range(1, 11):
        print(f"{i} cubed is {i**3}")

# Create two thread objects
t1 = threading.Thread(target=print_squares)
t2 = threading.Thread(target=print_cubes)

# Start the threads
t1.start()
t2.start()

# Wait for the threads to finish
t1.join()
t2.join()

print("Main thread exiting...")
```

```python
#Q5. State advantages and disadvantages of multithreading.

'''
Multithreading is a technique where multiple threads of execution run concurrently within a single process. Multithreading has 
several advantages and disadvantages:

Advantages of multithreading:

1. Improved performance: Multithreading can improve the performance of programs that involve multiple tasks, especially those that
   involve input/output operations or tasks that can be executed concurrently. By using multiple threads, a program can perform several tasks at once, which can result in faster execution times and improved overall performance.

2. Increased responsiveness: Multithreading can make a program more responsive to user input, as it allows the program to perform
   multiple tasks simultaneously while still being able to respond to user input in a timely manner.

3. Resource sharing: Threads can share resources such as memory and CPU time, which can lead to more efficient use of resources 
   and reduced memory usage.

Disadvantages of multithreading:

1. Complexity: Multithreading can make programs more complex and harder to debug, as it introduces concurrency and the possibility
   of race conditions and other synchronization issues.

2. Overhead: Multithreading introduces some overhead in terms of memory usage and CPU time. Creating and managing threads can be 
   resource-intensive, and too many threads can lead to decreased performance.

3. Synchronization issues: When multiple threads access shared resources, synchronization issues such as race conditions and 
   deadlocks can arise. These issues can be difficult to debug and can lead to unpredictable behavior in the program.

In summary, multithreading can provide significant performance benefits and increased responsiveness, but it also introduces 
complexity and the possibility of synchronization issues. When used appropriately, however, multithreading can be a powerful tool
for improving the performance and responsiveness of programs.
'''
pass
```
